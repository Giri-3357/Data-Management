### **Week 1 – Exercise 1 – Individual – Metadata Creation**

```python
# DataCite XML: Creates a DataCite metadata file for the Netflix dataset
datacite_xml = """<?xml version="1.0" encoding="UTF-8"?>
<resource xmlns="http://datacite.org/schema/kernel-4"
          xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
          xsi:schemaLocation="http://datacite.org/schema/kernel-4
          http://schema.datacite.org/meta/kernel-4/metadata.xsd">

    <identifier identifierType="DOI">10.1234/netflix.dataset.example</identifier>

    <creators>
        <creator>
            <creatorName>Shivam Bansal</creatorName>
        </creator>
    </creators>

    <titles>
        <title>Netflix Movies and TV Shows Dataset</title>
    </titles>

    <publisher>Kaggle</publisher>
    <publicationYear>2021</publicationYear>

    <resourceType resourceTypeGeneral="Dataset">Media and Entertainment Data</resourceType>

    <subjects>
        <subject>Entertainment</subject>
        <subject>Streaming Platforms</subject>
        <subject>Movies and Television</subject>
    </subjects>

    <descriptions>
        <description descriptionType="Abstract">
            This dataset provides an overview of movies and TV shows available on Netflix.
            It includes details such as title, director, cast, country, release year,
            rating, duration, and genre. It can be used for data analysis and recommendation systems.
        </description>
    </descriptions>

</resource>
"""

with open("netflix_datacite.xml", "w") as f:
    f.write(datacite_xml)

print("DataCite XML file created!")
```

```python
#schema.org JSON - Creates a JSON-LD metadata file for the Netflix dataset
import json

schemaorg_json = {
  "@context": "https://schema.org/",
  "@type": "Dataset",
  "name": "Netflix Movies and TV Shows Dataset",
  "description": "A dataset containing information about movies and TV shows available on Netflix.",
  "url": "https://www.kaggle.com/datasets/shivamb/netflix-shows",
  "creator": {
    "@type": "Person",
    "name": "Shivam Bansal"
  },
  "publisher": {
    "@type": "Organization",
    "name": "Kaggle"
  },
  "datePublished": "2021",
  "keywords": [
    "Netflix",
    "movies",
    "TV shows",
    "streaming",
    "entertainment"
  ]
}

with open("netflix_schemaorg.json", "w") as f:
    json.dump(schemaorg_json, f, indent=2)

print("Schema.org JSON file created!")
```
